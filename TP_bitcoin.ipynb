Enhancing Speed & Accuracy with Dimensionality Reduction
Problem Definition
You’ll predict a binary trading signal (1=buy, 0=sell) based on short vs. long moving averages and a set of technical indicators. Then you’ll reduce feature dimensionality (later: SVD, t-SNE) to (a) train faster, (b) avoid multicollinearity, and (c) visualize structure.
We cast a trading rule into a binary classifier:

```python
signal = 1 → Buy (short-term trend above long-term trend)
signal = 0 → Sell/No-buy (short-term ≤ long-term)
```

Setup 
More indicators ≠ better profits. Many correlated features bloat models, slow training, and overfit. DR compresses information into fewer, orthogonal components.
